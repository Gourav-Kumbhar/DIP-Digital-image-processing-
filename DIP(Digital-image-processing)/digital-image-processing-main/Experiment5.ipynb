# Image Sampling and Quantization

```python
# implement Image sampling and quantization to enhance the image

import numpy as np 
import cv2
import matplotlib.pyplot as plt

# read the image

original_image=cv2.imread('flower.jpg')[..., ::-1]
```

```python
# Image sampling (downsampling)
sampling_factor = 4  # Downsampling factor
sampled_image = original_image[::sampling_factor, ::sampling_factor]

# Image quantization
quantization_levels = 16  # Number of quantization levels (colors)
quantization_factor = 256 // quantization_levels
quantized_image = (original_image // quantization_factor) * quantization_factor


plt.imshow(original_image)
plt.title('Original Image')
plt.axis('off')

plt.show()
```

```python

plt.imshow(sampled_image)
plt.title('Sampled Image')
plt.axis('off')
plt.show()

```

```python
plt.imshow(quantized_image.astype(np.uint8))
plt.title('Quantized Image')
plt.axis('off')
plt.show()
```
