```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image=cv2.imread('vegetables.jpg')
colour_image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(colour_image)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray')
```

```python
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)
plt.imshow(blurred_image,cmap='gray')
```

```python
_, thresholded_image = cv2.threshold(gray_image, 128, 255, cv2.THRESH_BINARY)
plt.imshow(thresholded_image,cmap='gray')
```

```python
edges = cv2.Canny(gray_image, 10, 150)
plt.imshow(edges,cmap='gray')
```

```python
# Resize the image to a specific width and height
resized_image = cv2.resize(colour_image, (400, 300))
plt.imshow(resized_image)
```

```python
import numpy as np
import cv2

image = cv2.imread('flower.jpg')

height, width, _ = image.shape

# Create an empty grayscale image of the same size
gray_image = np.zeros((height, width), dtype=np.uint8)

# Iterate through each pixel in the color image and calculate the grayscale value
for i in range(height):
    for j in range(width):
        # Get the RGB values of the pixel
        b, g, r = image[i, j]
        
        gray_value = 0.299 * r + 0.587 * g + 0.114 * b
        
        # Set the corresponding pixel in the grayscale image
        gray_image[i, j] = int(gray_value)

# Save or display the grayscale image
cv2.imwrite('grayscale_image.jpg', gray_image)
cv2.imshow('Gray Image',gray_image)
```
